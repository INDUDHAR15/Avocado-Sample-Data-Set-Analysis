```python
my_list = []
N = int(input("Enter number"))

while N>0:
    command = input("Enter Command")
    N = N - 1
    if command.startswith("insert"):
        int_list = command.split()
        int_pos = int_list[1]
        int_val =int_list[2]
        my_list.insert(int(int_pos), int(int_val))
        #print(my_list)
    elif command == "print":
        print(my_list)
    elif command.startswith("remove"):
        split_list = command.split()
        rem_value = split_list[1]
        rum_value_int = int(rem_value)
        my_list.remove(rum_value_int)
        #print(my_list)
    elif command.startswith("append"):
        split_list = command.split()
        rem_value = split_list[1]
        rum_value_int = int(rem_value)
        my_list.append(rum_value_int)
        #print(my_list)   
    elif command == "sort":
        my_list.sort()
        #print(my_list)
    elif command == "pop":
        my_list.pop()
        #print(my_list)
    elif command == "reverse":
        my_list.reverse()
        #print(my_list)
    else:
        print("Not Valid")
```

```python
stu_list = [['Harry', 37.21], ['Berry', 37.21], ['Tina', 37.2], ['Akriti', 41], ['Harsh', 39]]

marks_set = set()
for x in stu_list:
    #print(x[1])
    marks = x[1]
    marks_set.add(marks)

marks_list = list(marks_set)

marks_list.sort()

sec_marks = marks_list[1]

#print(sec_marks)

red_list = []

for x in stu_list:
    if x[1] == sec_marks:
        red_list.append(x[0])

#print(red_list)
red_list.sort()

for i in red_list:
    print(i)
```
