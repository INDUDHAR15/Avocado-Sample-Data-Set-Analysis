```python
import numpy as np
import pandas as pd
import glob
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = r"Healthcare dataset"
```

```python
all_files = glob.glob(path + "/*.csv")
li = []
for filename in all_files:
    df = pd.read_excel(filename, index_col=None, header=0)
    li.append(df)
```

```python
# Filter data for a specific country, e.g., Afghanistan 
afghanistan_data = df[df['Entity'] == 'Afghanistan'] 
# Plot anxiety disorders over time 
plt.figure(figsize=(10, 6)) 
sns.lineplot(x='Year', y='Anxiety disorders (share of population) - Sex: Both - Age: Age-standardized', data=afghanistan_data) 
plt.title('Trends in Anxiety Disorders in Afghanistan') 
plt.xlabel('Year') 
plt.ylabel('Share of Population') 
plt.show()
```

```python
# Select relevant columns 
columns_of_interest = [ 'Year', 'Schizophrenia disorders (share of population) - Sex: Both - Age: Age-standardized', 
                       'Depressive disorders (share of population) - Sex: Both - Age: Age-standardized', 
                       'Anxiety disorders (share of population) - Sex: Both - Age: Age-standardized', 
                       'Bipolar disorders (share of population) - Sex: Both - Age: Age-standardized'
                      ]
```

```python
# Plot comparison 
plt.figure(figsize=(14, 8)) 
for column in columns_of_interest[1:]: 
    sns.lineplot(x='Year', y=column, data=df, label=column.split(' - ')[0])
plt.title('Comparison of Mental Health Disorders Over Time') 
plt.xlabel('Year') 
plt.ylabel('Share of Population') 
plt.legend() 
plt.show()
```
