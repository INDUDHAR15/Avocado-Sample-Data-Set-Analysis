```python
my_list = []
N = int(input("Enter number"))

while N>0:
    command = input("Enter Command")
    N = N - 1
    if command.startswith("insert"):
        int_list = command.split()
        int_pos = int_list[1]
        int_val =int_list[2]
        my_list.insert(int(int_pos), int(int_val))
        #print(my_list)
    elif command == "print":
        print(my_list)
    elif command.startswith("remove"):
        split_list = command.split()
        rem_value = split_list[1]
        rum_value_int = int(rem_value)
        my_list.remove(rum_value_int)
        #print(my_list)
    elif command.startswith("append"):
        split_list = command.split()
        rem_value = split_list[1]
        rum_value_int = int(rem_value)
        my_list.append(rum_value_int)
        #print(my_list)   
    elif command == "sort":
        my_list.sort()
        #print(my_list)
    elif command == "pop":
        my_list.pop()
        #print(my_list)
    elif command == "reverse":
        my_list.reverse()
        #print(my_list)
    else:
        print("Not Valid")
```

```python
def count_substring(string, sub_string, counter=0):
    if 1<= len(string) <= 200:
        found_index = string.find(sub_string)
        while found_index != -1:
            counter += 1
            found_index = string.find(sub_string,found_index+1)
        return counter

item_count = count_substring(str(input("Enter string")),str(input("Enter sub_string")))
print(item_count)
```

```python
s = input()
result = ""
for a in s:
    if a.isalnum():
        result = True
    else:
        result = False
print(f"{s} has any alphanumeric characters.", result)
for x in s:
    if x.isalpha():
        result = True
        break
    else:
        result = False
print(f"{s} has any alphabetical characters.", result)
for d in s:
    if d.isdigit():
        result = True
        break
    else:
        result = False
print(f"{s} has any digits characters.",result)
for l in s:
    if l.islower():
        result = True
        break
    else:
        result = False
print(f"{s} has any lowercase characters.",result)
for u in s:
    if u.isupper():
        result = True
        break
    else:
        result = False
print(f"{s} has any uppercase characters.",result)
```

```python
#Replace all ______ with rjust, ljust or center. 

thickness = int(input()) #This must be an odd number
c = 'H'

#Top Cone
for i in range(thickness):
    print((c*i).rjust(thickness-1)+c+(c*i).ljust(thickness-1))

#Top Pillars
for i in range(thickness+1):
    print((c*thickness).center(thickness*2)+(c*thickness).center(thickness*6))

#Middle Belt
for i in range((thickness+1)//2):
    print((c*thickness*5).center(thickness*6))    

#Bottom Pillars
for i in range(thickness+1):
    print((c*thickness).center(thickness*2)+(c*thickness).center(thickness*6))    

#Bottom Cone
for i in range(thickness):
    print(((c*(thickness-i-1)).rjust(thickness)+c+(c*(thickness-i-1)).ljust(thickness)).rjust(thickness*6))
```

```python
import textwrap

def wrap(string, max_width):
    s_split = textwrap.wrap(string, max_width)
    splited_str = "\n".join(s_split)
    print(splited_str)

wrap("ABCDEFGHIJKLIMNOQRSTUVWXYZ",4)
```
