```python
s = input()
result = ""
for a in s:
    if a.isalnum():
        result = True
    else:
        result = False
print(f"{s} has any alphanumeric characters.", result)
for x in s:
    if x.isalpha():
        result = True
        break
    else:
        result = False
print(f"{s} has any alphabetical characters.", result)
for d in s:
    if d.isdigit():
        result = True
        break
    else:
        result = False
print(f"{s} has any digits characters.",result)
for l in s:
    if l.islower():
        result = True
        break
    else:
        result = False
print(f"{s} has any lowercase characters.",result)
for u in s:
    if u.isupper():
        result = True
        break
    else:
        result = False
print(f"{s} has any uppercase characters.",result)
```

```python
def minion_game(string):
    # your code goes here
    vowel = 'AEIOUaeiou' 
    score_kevin = 0 
    score_stuart = 0

    for i in range(len(string)):
        if string[i] in vowel:
            score_kevin += len(string) - i 
        else:
            score_stuart += len(string) - i  

    if score_stuart > score_kevin:
        print('Stuart', score_stuart)
    elif score_kevin > score_stuart:
        print('Kevin', score_kevin)
    else:
        print('Draw')

minion_game("Money eating BANANA's")
```

3. Write a program Quadratic to find the roots of the equation (a* x * x + b * x + c). Since
the equation is x*x, hence there are 2 roots. The 2 roots of the equation can be
found using a formula

delta = b * b - 4 * a * c

Root 1 of x = (-b + sqrt(delta))/(2*a)

Root 2 of x = (-b - sqrt(delta))/(2*a)

Take a, b, and c as input values to find the roots of x.

```python
import math
a, b, c = int(input()), int(input()), int(input())

delta = ((b*b) - (4*a*c))
print(delta)
root1 = (-b + math.sqrt(delta))/(2*a)
print(root1)
root2 = (-b - math.sqrt(delta))/(2*a)
print(root2)
```

4. Write a TemperatureConversion program, given the temperature in Fahrenheit as
input outputs the temperature in Celsius or vice versa using the formula

Celsius to Fahrenheit: (°C × 9/5) + 32 = °F

Fahrenheit to Celsius: (°F − 32) x 5/9 = °C.

```python
n = int(input("Enter Temperature in next input enter 'C' for Celsius and 'F' for Fahrenheit."))
cel_fah = str(input("Enter C or F"))

C = ((n * (9/5)) + 32)
F = ((n - 32 * (5/9)))
if cel_fah == "C":
    print(n,cel_fah, "=",C,"Fahrenheit")
elif cel_fah == "F":
    print(n,cel_fah, "=",F, "Celsius")
else:
    print("Wrong Enter")
```
